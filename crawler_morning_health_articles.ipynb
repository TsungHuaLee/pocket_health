```python
import requests
from bs4 import BeautifulSoup

base_url = "https://www.edh.tw"

# 開始爬蟲
def start_crawling(url, pages_we_want, output_file):
    
    for counter in range(pages_we_want):
        print("=====================================================")
        print("這一頁的url: " + url)
        
        r = requests.get(url)
        soup = BeautifulSoup(r.text, "html.parser")

        # get title
        title_sel = soup.select(".title a")

        # get abstract
        abstract_sel = soup.select(".list-article a.content")

        # get thumbnail
        thumbnail_sel = soup.select(".list-img img")

        # check array size
        print("abstract array size: " + str(len(abstract_sel)) + ", title size: " + str(len(title_sel)) + ", thumbnail size: " + str(len(thumbnail_sel)))

        # get page info
        for index, s in enumerate(title_sel):
            if index < 10:
                write_csv(s.text, str(abstract_sel[index].text).strip(), base_url + s["href"], str(thumbnail_sel[index]["src"]))

        # next page
        nextpage_sel = soup.select(".pagination li")
        url = nextpage_sel[len(nextpage_sel)-2].a["href"]

    output_file.close()
    print("===== 爬蟲結束 =====")
```

# 糖尿病，有圖片，存csv

```python
import requests
from bs4 import BeautifulSoup
import csv

base_url = "https://www.edh.tw"

# get all pages
url = "https://www.edh.tw/tag/%E9%81%8B%E5%8B%95/1"
pages_we_want = 296
output_file = open('運動_早安健康.csv', 'w', newline='', encoding="utf_8_sig")

# 建立 CSV 檔寫入器
writer = csv.writer(output_file)

# 寫入一列資料
writer.writerow(['文章標題', '摘要', '文章url', '縮圖url'])

def write_csv(title, abstract, url, thumbnail_url):
    # 寫入列資料
    writer.writerow([title, abstract, url, thumbnail_url])
    
    # 輸出
    print("url: " + base_url + s["href"] + ", title: " + s.text + ", abstract: " + str(abstract_sel[index].text).strip() + ", img: " + str(thumbnail_sel[index]["src"]))
    print("--------------------")

for counter in range(pages_we_want):
    print("=====================================================")
    print("這一頁的url: " + url)
    r = requests.get(url)
    soup = BeautifulSoup(r.text, "html.parser")

    # get title
    title_sel = soup.select(".title a")

    # get abstract
    abstract_sel = soup.select(".list-article a.content")
    
    # get thumbnail
    thumbnail_sel = soup.select(".list-img img")

    # check array size
    print("abstract array size: " + str(len(abstract_sel)) + ", title size: " + str(len(title_sel)) + ", thumbnail size: " + str(len(thumbnail_sel)))
    
    # get page info
    for index, s in enumerate(title_sel):
        if index < 10:
            write_csv(s.text, str(abstract_sel[index].text).strip(), (base_url + s["href"]), str(thumbnail_sel[index]["src"]))
        
    # next page
    nextpage_sel = soup.select(".pagination li")
    url = nextpage_sel[len(nextpage_sel)-2].a["href"]

output_file.close()
print("===== 爬蟲結束 =====")
```
