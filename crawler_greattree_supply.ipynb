```python
import requests
from bs4 import BeautifulSoup
import csv

base_url = "https://shop.greattree.com.tw/greattree/"

# init url
url = "https://shop.greattree.com.tw/greattree/index.php?action=product_sort&prod_sort_uid=599"

# get max page num
r = requests.get(url)
soup = BeautifulSoup(r.text, "html.parser")
max_page_sel = soup.select(".Kg_pageBox .countlist")
pages_we_want = int(max_page_sel[len(max_page_sel)-1].text)

# init variables
category = '營養補充'
output_file = open('保健食品_' + category + '_大樹.csv', 'w', newline='', encoding="utf_8_sig")

# 建立 CSV 檔寫入器
writer = csv.writer(output_file)

# 寫入一列資料
writer.writerow(['商品分類', '商品標題', '原價格', '特價', '商品url', '縮圖url'])

def write_csv(title, origin_price, onsale_price, url, thumbnail_url):
    # 寫入列資料
    writer.writerow([category, title, origin_price, onsale_price, url, thumbnail_url])
    
    # 輸出
    print("title: " + title + ", origin_price: " + origin_price + ", onsale: " + onsale_price + ", url: " + url + ", thumbnail: " + thumbnail_url)
    print("--------------------")

for counter in range(pages_we_want):
    print("=====================================================")
    print("這一頁的url: " + url)

    # get title
    title_sel = soup.select(".search-product-title a")
    print(title_sel[0].text)

    # get origin price
    origin_price_sel = soup.select(".o_price")
    
    # get onsale price
    onsale_price_sel = soup.select(".price_color")
    
    # get item url
    item_url_sel = soup.select(".search-product-photo a")
    
    # get thumbnail
    thumbnail_sel = soup.select(".search-product-photo a img")

    # check array size
    print("title size: " + str(len(title_sel)) + ", origin price size: " + 
          str(len(origin_price_sel)) + ", onsale price size: " + 
          str(len(onsale_price_sel)) + ", thumbnail size: " + str(len(thumbnail_sel)))
    
    # get page info
    for index, s in enumerate(title_sel):
        end = onsale_price_sel[index].text[1:].index("$")
        write_csv(s.text, 
                  str(origin_price_sel[index].text), 
                  str(onsale_price_sel[index].text[:end+1]), 
                  str(item_url_sel[index]["href"]), 
                  str(thumbnail_sel[index]["data-original"]))
        
    # next page
    nextpage_sel = soup.select(".next a")
    url = base_url + nextpage_sel[0]["href"]

output_file.close()
print("===== 爬蟲結束 =====")
```
